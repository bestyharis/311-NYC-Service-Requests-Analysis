```python
df.drop(['School or Citywide Complaint','Vehicle Type','Taxi Company Borough','Taxi Pick Up Location','Garage Lot Name'],inplace=True,axis=1)
```

```python
# Q2 Read or convert the columns ‘Created Date’ and Closed Date’ to datetime datatype and create 
# a new column ‘Request_Closing_Time’ as the time elapsed between request creation and request closing

df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
df['Request_Closing_Time'] =(df['Closed Date'] - df['Created Date'])
df['Request_Closing_Time_Minutes'] =(df['Closed Date'] - df['Created Date']).dt.seconds/60
```

```python
agency_complaint = df.groupby("Agency")["Complaint Type"].value_counts().to_frame().rename(columns = {'Complaint Type':'Count'}).sort_values('Count',ascending = False)
```

```python
agency_complaint
agency_complaint.reset_index(inplace=True)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,15))
plt.style.use('seaborn-pastel')


plt.grid()
plt.barh(agency_complaint['Complaint Type'],agency_complaint['Count'],
         label = 'Complaint Count',color = 'green', edgecolor = 'black',linewidth = 1.2,alpha = 0.8);

for i,v in enumerate(agency_complaint['Count']):
    plt.text(v+0.2,i+0.1,str(v), color = 'blue',fontweight = 'bold')

plt.show()

#Maximum number of the Complaints are from Blocked Driveway and Minimum are from Animal in the Park
```

```python
complain_descriptor = df.groupby(['City','Descriptor'])['Complaint Type'].count().to_frame().sort_values('Complaint Type',ascending = False)
complain_descriptor.reset_index(inplace = True)

# Top 15 Complaint's

complain_descriptor2 = complain_descriptor.iloc[:15]

plt.pie(complain_descriptor2['Complaint Type'], labels= complain_descriptor2['Descriptor'],
       autopct='%0.1f%%',radius = 2,textprops={'fontsize':15,'color':'black'})
plt.show()
complain_descriptor2
```

```python
city_complaints = df.groupby('City')['Complaint Type'].count().to_frame().sort_values('Complaint Type',ascending = False)
city_complaints.reset_index(inplace = True)

# Top 15 Complaint's

city_complaints2 = city_complaints.iloc[:5]
plt.pie(city_complaints2['Complaint Type'], labels= city_complaints2['City'],autopct='%0.1f%%',radius = 2,shadow = True,textprops={'fontsize':15,'color':'black'})
plt.show()

city_complaints2
```

```python
df['Request Closing Time in Seconds'] = df['Request_Closing_Time'].apply(lambda x:x.seconds)
```

```python
from scipy.stats import ttest_ind
average_response_time = {}
unique_complaint = set(df['Complaint Type'].unique())

for i in unique_complaint:
    average_response_time[i] =df[df['Complaint Type']==i]['Request Closing Time in Seconds'].mean()
```

```python
for i in unique_complaint:
    for k in set(unique_complaint)-{i,i}:
        print("mean for {} is - {}".format(i,average_response_time[i]))
        print("mean for {} is - {}".format(k,average_response_time[k]))
        
        set1 = list(df[df['Complaint Type']==i]['Request Closing Time in Seconds'].dropna())
        set2 = list(df[df['Complaint Type']==k]['Request Closing Time in Seconds'].dropna())
        
        ttest_stat,p_value = ttest_ind(set1,set2)
        print("{} p values for similar average for {},{}\n\n".format(round(p_value,3),i,k))
```
